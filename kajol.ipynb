#1
matrix of A with order(n,k) and matrix of b with order (k,m).
let c be the multiplication of matrix a and matrix b with order(n,m)
in each element in matrix c has we have to do k multiplications and k-1 additions
so,matrix c has n*m elements 
therefore, the total number of multiplications are k*m*n
          total number of additions are (k-1)*m*n

```python
#2
import numpy as np
import time

def matrix_multiply(A, B):
    n = len(A)     
    k = len(B) 
    m = len(B[0])  
    C = [[0] * m for _ in range(n)]
    for i in range(n):
        for j in range(m):
            C[i][j] = sum(A[i][l] * B[l][j] for l in range(k))
    
    return C
def numpy_dot(A,B):
     return np.dot(A,B)




A=[ [2,3,4],
   [5,6,7]
  ]
b=[[1,2],
   [3,4],
   [5,6]
  ]
start_time = time.time()
manual_result =matrix_multiply(A,b)
end_time = time.time()
print("Time taken for manual dot product:", end_time - start_time)
start_time = time.time()
numpy_result = numpy_dot(A,b)
end_time = time.time()
print("Time taken for NumPy dot product:", end_time - start_time)
print( matrix_multiply(A,b))
print( numpy_dot(A,b))
```

```python
#3
import numpy as np
import time

#normal method
def pass_median(A):
    n = len(A)
    for i in range(n // 2 + 1):
        max_i = i
        for j in range(i + 1, n):
            if A[j] > A[max_i]:
                max_i = j
        A[i], A[max_i] = A[max_i], A[i]
    if n % 2 == 1:
        return A[n // 2]
    else:
        return (A[n // 2 - 1] + A[n // 2]) / 2

#quick select method
def quickselect_median(A):
    n = len(A)
    if n % 2 == 1:
        return quickselect(A, n // 2)
    else:
        return (quickselect(A, n // 2 - 1) + quickselect(A, n // 2)) / 2

def quickselect(A, k):
    if len(A) == 1:
        return A[0]
    pivot_i = len(A)//2
    pivot = A[pivot_i]
    left = [x for i, x in enumerate(A) if x < pivot and i != pivot_i]
    middle = [x for x in A if x == pivot]
    right = [x for i, x in enumerate(A) if x > pivot and i != pivot_i]
    if k < len(left):
        return quickselect(left, k)
    elif k < len(left) + len(middle):
        return middle[0]
    else:
        return quickselect(right, k - len(left) - len(middle))
A=[12,11,9,34,0,96]
#for pass median time complexity is O(n^2)
#for quick select time complexity is O(n)
# for numpy time complexity is O(n logn)
# so,quick select is best
```

gradient w.r.t y is x^2+3y^2 sin(x)

```python
#6
import sympy as sp


# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = (x**2)*y+(y**3)*(sp.sin(x))

# Compute the gradient analytically
grad_f = sp.Matrix([f.diff(x), f.diff(y)])

print(grad_f)
```

```python
#10
import numpy as np
import pandas as pd


np.random.seed(0) 
matrix = np.random.uniform(1, 2, size=(20, 5))

columns = ["a", "b", "c", "d", "e"]
df = pd.DataFrame(matrix, columns=columns)


stddev = df.std()  
max_std_column = stddev.idxmax() 
max_std_value = stddev[max_std_column]  

print(f"Column with the highest standard deviation: {max_std_column}")
print(f"Standard deviation value: {max_std_value}")


means = df.mean(axis=1)  
min_mean_row_index = means.idxmin()  
min_mean_value = means[min_mean_row_index]  

print(f"Row with the lowest mean: {min_mean_row_index}")
print(f"Mean value: {min_mean_value}")
```

```python
#11
import numpy as np
import pandas as pd


np.random.seed(0) 
matrix = np.random.uniform(1, 2, size=(20, 5))


columns = ["a", "b", "c", "d", "e"]
df = pd.DataFrame(matrix, columns=columns)


df['f'] = df.sum(axis=1)


df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')


num_LT8 = (df['g'] == 'LT8').sum()


stddev_LT8 = df[df['g'] == 'LT8']['f'].std()
stddev_GT8 = df[df['g'] == 'GT8']['f'].std()

# Print results
print(f"Number of rows where 'g' is 'LT8': {num_LT8}")
print(f"Standard deviation of 'f' for rows where 'g' is 'LT8': {stddev_LT8}")
print(f"Standard deviation of 'f' for rows where 'g' is 'GT8': {stddev_GT8}")
```

```python
#13
import numpy as np
arr=np.array([24,35,46,57,68,12,79,80])

def min(arr):
    min_val = arr[0]
    min_i = 0
    for i in range(1, len(arr)):
        if arr[i] < min_val:
            min_val = arr[i]
            min_i = i
    return min_i
print(np.argmin(arr))
print(min(arr))
```
